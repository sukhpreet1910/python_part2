```python
import requests, json


def get_data(id):
    url = f'https://pokeapi.co/api/v2/pokemon/{id}/'
    response = requests.get(url)
    return response.json()

data = []
for i in range(1,100):
    data.append(get_data(i))

    
```

```python
def types(data):
    all_types = []
    for types in data:
        for poke_type in types['types']:
            curr_type = poke_type['type']['name'] 
            if curr_type not in all_types:
                all_types.append(curr_type)
    return all_types
    
    
all_types = types(data)
print(all_types)
```

```python
def organized_data(data):
    final_data = []
    
    for curr_type in all_types:
        poke_dict = {}
        poke_dict[curr_type] = {}  
        for poke in data:
            
            poke_type = [types['type']['name'] for types in poke['types']]
            poke_name = poke['name']

            if curr_type in poke_type:
                
                poke_dict[curr_type][poke_name] = {
                    'abilities': [ability['ability']['name'] for ability in poke['abilities']],
                    'weight': poke['weight'],
                }
               
                if poke_dict not in final_data:
                    final_data.append(poke_dict)
    return final_data


result = organized_data(data)
result
# for r in result:
#     print(r, '\n')
```
