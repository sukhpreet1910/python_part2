```python
sorted_data = []

def get_info(data):

    for racer in data:

        racer_dict = {}
        racer_name = f"{racer['Driver']['givenName']} {racer['Driver']['familyName']}"

        racer_dict[racer_name] = {'position': racer['position'],
                                 'Code': racer['Driver']['code'],
                                 'first_name': racer["Driver"]['givenName'],
                                 'last_name': racer["Driver"]['familyName'],
                                 'wins': racer['wins'],
                                 'Nationality': racer["Driver"]['nationality'],
                                 'car_name': racer['Constructors'][0]['name'],
                                 'company_nationality': racer['Constructors'][0]['nationality'],
                    }

        sorted_data.append(racer_dict)
    return sorted_data
    
```

```python


def get_info_by_year_and_round(year, r):
    
    url = f'http://ergast.com/api/f1/{year}/{r}/driverStandings.json'
    
    response = requests.get(url)
    if not response.json()['MRData']['StandingsTable']['StandingsLists']:
        return "We Had An Arror Loading DataðŸ’© ðŸ’©"
    
    
    
    year_raw_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
    return get_info(year_raw_data)
    

year = int(input("Enter the year, you want the data for: "))
rounds = int(input("Enter number of round: "))
get_info_by_year_and_round(year, rounds)
```
